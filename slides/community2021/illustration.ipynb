```python
def project_name(key):
    parts = key.split("-")
    if parts[0] == "incubator" and len(parts)>1:
        return parts[1]
    return parts[0]
prs = pd.read_csv("prs.csv")
prs = prs.rename(columns={"count(*)":"pr"})
prs["project"] = prs.repo.apply(project_name) 
```

```python
prs.sort_values("pr", ascending=False)
x = prs.groupby("project").sum().reset_index().sort_values("pr", ascending=False)
```

```python
s = pd.read_csv("spark90.csv")
fig1, ax1 = plt.subplots()
ax1.pie(s["pr"],shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(s["contribution"],shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```
